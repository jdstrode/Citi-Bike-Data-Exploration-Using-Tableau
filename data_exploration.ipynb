```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# File to Load (Remember to Change These)
Data_Jun_2020 = "202006-citibike-tripdata.csv"
Data_May_2020 = "202005-citibike-tripdata.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
df_Jun_2020 = pd.read_csv(Data_Jun_2020, encoding="utf-8")
df_May_2021 = pd.read_csv(Data_May_2020, encoding="utf-8")
```
